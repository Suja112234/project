```python
import pandas as pd
import numpy as np

from sklearn.impute import KNNImputer

df=pd.read_csv('C:/Users/DELL/Downloads/life_expectancy.csv')
df.head()
```

```python
#Sanity Check
print(df.shape)
print(df.size)
print(df.info())
```

```python
print(df.describe())
```

```python
df.head()
```

```python
df.drop(columns=['id'], inplace=True)
df.columns
```

```python
# Handling Duplicates
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
print(df.duplicated().sum())
```

```python
#Handling missing values
df.isnull().sum()
```

```python
#Dropping Rows
df.dropna(subset=['life_expectancy'], inplace=True)
df.isnull().sum()
```

```python
#Filling
df['adult_mortality'] = df['adult_mortality'].fillna(0)
print(df.isnull().sum())
```

```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
print(df.isnull().sum())
```

```python
cols_for_mean = ['bmi', 'total_expenditure']
for col in cols_for_mean:
    df[col] = df[col].fillna(df[col].mean())
df.isnull().sum()
```

```python
#Imputing using KNN Imputer
cols_for_impute = ['gdp', 'population', 'schooling']
ki = KNNImputer()
for col in cols_for_impute:
    df[col] = ki.fit_transform(df[[col]])
df.isnull().sum()
```

```python
#Label Encoding
df.head()
```

```python
df['country'] = pd.factorize(df['country'])[0]
df.head()
```

```python
df['status'].unique()
```

```python
df = pd.get_dummies(data=df, columns=['status'], drop_first=True)
df.head()
```
