```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
tips = sns.load_dataset('tips')

# Basic information
print(tips.head())
print(tips.info())
print(tips.describe())

# Univariate Analysis
sns.histplot(tips['total_bill'], kde=True)
plt.title('Distribution of Total Bill')
plt.show()

sns.boxplot(x=tips['total_bill'])
plt.title('Box Plot of Total Bill')
plt.show()

# Bivariate Analysis
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip')
plt.show()

sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Total Bill by Day')
plt.show()

# Identify categorical columns
categorical_cols = tips.select_dtypes(include=['object', 'category']).columns
print("Categorical Columns:", categorical_cols)

# Example: Convert 'sex' column to numerical
# If you need to convert the 'sex' column into numerical format, you can use mapping
tips['sex_num'] = tips['sex'].map({'Male': 0, 'Female': 1})

# Now you can perform numerical operations
sns.scatterplot(x='total_bill', y='tip', hue='sex_num', data=tips)
plt.title('Total Bill vs Tip (Colored by Sex)')
plt.show()

# Alternatively, you can exclude non-numeric columns when plotting
# For instance, when using sns.pairplot()
numeric_cols = tips.select_dtypes(include=['number']).columns
sns.pairplot(tips[numeric_cols])
plt.show()
```
