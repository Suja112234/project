```python
agg_df = df[['sepal_length', 'petal_length', 'species']].groupby('species').mean().reset_index()
# sns.barplot(x=agg_df.index, y=agg_df['petal_length'], hue=agg_df['species'])
```

```python
import matplotlib.pyplot as plt
```

```python
# Bar plot of Product Category counts
plt.figure(figsize=(15, 4))
sns.countplot(x='petal_length', data=df)
plt.title('Petal Length Sand species')
plt.xlabel('petal_length')
plt.ylabel('species')
plt.show()
```

```python
sns.barplot(x=agg_df['species'], y=agg_df['sepal_length'])
sns.barplot(x=agg_df['species'], y=agg_df['petal_length'])
```

```python
sns.barplot(x=df['species'], y=df['petal_length'])
```

Pie Chart

```python
# plt.pie(iris.sepal_length, labels=iris.species)
# plt.show()
pivot_table = pd.pivot_table(df[["sepal_length", "species"]], index="species", aggfunc="sum")
plt.pie(pivot_table["sepal_length"], labels=pivot_table.index, explode = (0,0,0.05), autopct="%0.1f%%")
plt.show()
```
